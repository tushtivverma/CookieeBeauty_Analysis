```python
# Converting Time & Date to datetime64
df['Time & Date'] = pd.to_datetime(df['Time & Date'], errors='coerce')
```

```python
# Filling NaN values with 0
df['Shipping'] = df['Shipping'].fillna(0)
```

```python
# Filling NaN values with 0
df['Discount'] = df['Discount'].fillna(0)
```

```python
# Converting columns from object to int64
df['Amount'] = df['Amount'].str.replace(',','').astype('float64').round().astype('int64')
df['Discount'] = df['Discount'].str.replace(',','').astype('float64').round()
```

```python
df.info()
```

```python
#No requirement for the OrderType Column
df = df.drop(columns = ['Order Type','Payment Type', 'Currency'])
```

```python
#Added a new column to sum up the total amount of an order
df['Total Amount'] = df['Amount'] - df['Discount'] + df['Shipping']
```
